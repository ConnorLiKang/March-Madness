```python
#linear_model.Ridge(alpha=0.5)
#neuralNetwork(10)
#AdaBosstClassifier(n_estimators = 100)
#RandomForestClassifier(n_estimators =64 )
import pandas as pd
import numpy as np
df = pd.read_csv('shuffle.csv')
df.head()
```

```python
#define columns value into input x and output y
array = df.values
x = array[:,0:66]
y = array[:,66]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import cross_validation
sc = StandardScaler()
x_std = sc.fit_transform(x)
pca = PCA(n_components = 'mle',svd_solver='full')
x_pca = pca.fit_transform(x_std)
x_train_pca, x_test_pca, y_train, y_test = cross_validation.train_test_split(x_pca,y,test_size=0.3,random_state = 42)
x_train_pca.shape,x_test_pca.shape
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV


NN = MLPClassifier()
param_gridNN = [{#'hidden_layer_sizes':[(20,10),(50,10)],
               'activation':['logistic','tanh','relu'],
               'solver':['lbfgs','sgd','adam'],
               'alpha':[1e-3,1e-2,1e-1,1],
                'max_iter':[100000]
               #'leanrning_rate': ['constant','invascaling','adaptive']
              }]
               
#                    'solver':['liblinear'],
#                    'multi_class':['ovr']},
#                {'penalty':['l2'],
#                 'C':[0.01,0.05,0.1,0.5,1,5,10,50,100],
#                'solver':['lbfgs'],
#                'multi_class':['ovr','multinomial']}]
search = GridSearchCV(NN, param_gridNN, cv=10)
search.fit(x_pca, y)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)
```

```python
from sklearn.metrics import accuracy_score

y_pred = search.predict(x_pca)
accuracy = accuracy_score(y, y_pred)
accuracy
```

```python
#返回probability 第二列表示该场比赛team1 win的概率
y_prob = search.predict_proba(x_pca)
y_prob
```

```python
df['team1_outcome_pred'] = y_pred[:]
team1_win = []
for lst in y_prob:
    p = lst[1]
    team1_win.append(p)
team1_win
```

```python
df['probability'] = team1_win
df.head()
```
